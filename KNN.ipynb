```python
En este ejercicio, utilizaré el Iris Flower Dataset para realizar una carga, exploración básica y modelado utilizando el algoritmo
de K-Nearest Neighbors (KNN)
```

```python
#Conjunto de Datos: Iris FlowerDataset.
from sklearn.datasets import load_iris
iris = load_iris()

#Guardar el DataFrame de pandas
import pandas as pd
X = pd.DataFrame(iris.data, columns=iris.feature_names)
y = pd.Series(iris.target, name='target')

# Descripción del conjunto de datos
print(iris.DESCR)

# Resumen estadístico de las características.
print(X.describe())
print(X.info())

#Entrenamiento
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)


# Clasificador K-NearestNeighbors con k=1

from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=1)
knn.fit(X_train, y_train)

# Rendimiento del clasificador en el conjunto de prueba.
from sklearn.metrics import accuracy_score
y_pred = knn.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print('Accuracy:', accuracy)



#Tasa de error (misclassificationrate)
misclassification_rate = 1 - accuracy
print('Misclassification rate:', misclassification_rate)

#Precisión del modelo.
from sklearn.metrics import precision_score
precision = precision_score(y_test, y_pred, average='weighted')
print('Precision:', precision)

# Matriz de confusión
from sklearn.metrics import confusion_matrix
confusion_matrix = confusion_matrix(y_test, y_pred)
print('Confusion matrix:\n', confusion_matrix)

from sklearn.metrics import ConfusionMatrixDisplay, confusion_matrix
import matplotlib.pyplot as plt

# Visualización
disp = ConfusionMatrixDisplay(confusion_matrix=cm)
k = 1
disp.plot(cmap='Blues')
plt.title(f'Matriz de Confusión (k={k})')
plt.show()
```
